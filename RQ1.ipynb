```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from scipy import stats
```

# Figure 4: City-level Scaling

```python
restaurant_df = pd.read_csv('data/Chinese_city_restaurant.csv')
```

```python
# Total O2O Food Delivery Restaurants in Chinese Cities
df = restaurant_df.copy()
column = 'poi_num'

df['population'] = df['population'] * 10000
city_num = df['city_name'].nunique()
tmpdf = df[['population', column]].dropna()
logdf = np.log10(tmpdf)

fig, ax = plt.subplots(figsize=(6.5, 5.5))
logdf['intercept'] = 1 
model = sm.OLS(logdf[column], logdf[['intercept', 'population']])
results = model.fit()
intercept, slope = results.params

conf_interval = results.conf_int().loc['population']
plt.plot(logdf['population'], intercept + slope * logdf['population'], color='red', linestyle='-')

plt.scatter(logdf['population'], logdf[column], alpha=0.5, s = 30, color = 'gray')
plt.xlabel('Population', fontsize=25)
plt.ylabel('#Restaurants', fontsize=25)

plt.xticks([5, 6, 7], ['10$^5$', '10$^6$', '10$^7$'], fontsize = 20)
plt.yticks([2, 3, 4, 5], ['10$^2$', '10$^3$', '10$^4$', '10$^5$'], fontsize = 20)
plt.tight_layout()
plt.savefig('H://2024秋/ICWSM_food_delivery/figs/scaling1.pdf', transparent = True, dpi = 300)

plt.show()

print(f'Fitted Equation: log(y) = {intercept:.3f} + {slope:.3f} * log(x)')
print(f'95% Conf. Interval of Scaling Coefficient: [{conf_interval[0]:.3f}, {conf_interval[1]:.3f}]')
print(f'R^2: {results.rsquared:.3f}')
```

# Figure 5: Individual Level Food Choice Discrepancy

```python
user_demog = pd.read_csv('data/user_demog.csv')
user_pay = pd.read_csv('data/user_total_payment.csv')
user_preference = pd.read_csv('data/user_ff_preference.csv')
```

```python
tot_col = [x for x in user_preference.columns if 'total' in x]
ff_col = [x for x in user_preference.columns if 'ff' in x and 'nff' not in x]
user_preference['ff_ratio'] = user_preference[ff_col].sum(axis = 1) / user_preference[tot_col].sum(axis = 1)
user_preference['tot_orders'] = user_preference[tot_col].sum(axis = 1)

pays1 = pd.merge(user_pay, user_preference[['user_code', 'ff_ratio', 'tot_orders']])
pays1['avg_price'] = pays1.total_pay / pays1.tot_orders
pays1 = pd.merge(pays1, user_demog[['userid', 'gender_label', 'age_label', 'salary']], left_on='user_code', right_on='userid', how='left')
```

```python
c1 = '#6384B3'
c2 = '#F08743'
```

```python
desired_order = ['<25 yr', '25~40 yr', '>40 yr']

mean_ff_ratio = pays1[pays1.age_label != 'Other'].groupby('age_label')['ff_ratio'].mean().reindex(desired_order)
mean_avg_price = pays1[pays1.age_label != 'Other'].groupby('age_label')['avg_price'].mean().reindex(desired_order) / 2

ci_ff_ratio = pays1[pays1.age_label != 'Other'].groupby('age_label')['ff_ratio'].agg(lambda x: stats.t.interval(0.95, len(x)-1, loc=x.mean(), scale=stats.sem(x))[1]).reindex(desired_order)
ci_avg_price = pays1[pays1.age_label != 'Other'].groupby('age_label')['avg_price'].agg(lambda x: stats.t.interval(0.95, len(x)-1, loc=x.mean(), scale=stats.sem(x))[1]).reindex(desired_order) / 2

fig, ax1 = plt.subplots(figsize = (9, 7.5))
bars1 = ax1.bar(np.arange(mean_ff_ratio.shape[0]) - 0.2, mean_ff_ratio, width = 0.35, yerr= ci_ff_ratio - mean_ff_ratio, capsize=10, color=c1)
ax1.set_ylabel('Fast Food Order Ratio', color=c1, size = 35)
ax1.tick_params(axis='y', labelcolor=c1, labelsize=30)
ax2 = ax1.twinx()
bars2 = ax2.bar(np.arange(mean_ff_ratio.shape[0]) + 0.2, mean_avg_price, width = 0.35, yerr=ci_avg_price - mean_avg_price, capsize=10, color=c2)
ax2.set_ylabel('Average Order Price (CNY)', color=c2, size = 35)
ax2.tick_params(axis='y', labelcolor=c2, labelsize=30)
ax1.set_xlabel('Age', size = 35)
ax1.set_xticks([0, 1, 2], desired_order, size = 30)
plt.tight_layout()
plt.show()
```

```python
desired_order = ['Female', 'Male']

mean_ff_ratio = pays1[pays1.gender_label != 'Other'].groupby('gender_label')['ff_ratio'].mean().reindex(desired_order)
mean_avg_price = pays1[pays1.gender_label != 'Other'].groupby('gender_label')['avg_price'].mean().reindex(desired_order) / 2

ci_ff_ratio = pays1[pays1.gender_label != 'Other'].groupby('gender_label')['ff_ratio'].agg(lambda x: stats.t.interval(0.95, len(x)-1, loc=x.mean(), scale=stats.sem(x))[1]).reindex(desired_order)
ci_avg_price = pays1[pays1.gender_label != 'Other'].groupby('gender_label')['avg_price'].agg(lambda x: stats.t.interval(0.95, len(x)-1, loc=x.mean(), scale=stats.sem(x))[1]).reindex(desired_order) / 2

fig, ax1 = plt.subplots(figsize = (9, 7.5))
bars1 = ax1.bar(np.arange(mean_ff_ratio.shape[0]) - 0.2, mean_ff_ratio, width = 0.35, yerr= ci_ff_ratio - mean_ff_ratio, capsize=10, color=c1)
ax1.set_ylabel('Fast Food Order Ratio', color=c1, size = 35)
ax1.tick_params(axis='y', labelcolor=c1, labelsize=30)
ax2 = ax1.twinx()
bars2 = ax2.bar(np.arange(mean_ff_ratio.shape[0]) + 0.2, mean_avg_price, width = 0.35, yerr=ci_avg_price - mean_avg_price, capsize=10, color=c2)
ax2.set_ylabel('Average Order Price (CNY)', color=c2, size = 35)
ax2.tick_params(axis='y', labelcolor=c2, labelsize=30)
ax1.set_xlabel('Gender', size = 35)
ax1.set_xticks([0, 1], desired_order, size = 30)
plt.tight_layout()
plt.show()
```
